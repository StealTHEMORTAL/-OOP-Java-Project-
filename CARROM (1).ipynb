```python
import SimpleGUICS2Pygame.simpleguics2pygame as simplegui
import math

# Globals
pocket_coordinates = [(42, 42), (42, 558), (558, 42), (558, 558)]
pocket_radius = 25
width = 600
carrom_width = 570
carrom_height = 570
height = 600
half_w = width // 2
half_h = height // 2
c_rad = 18
player_score = 0
striker_radius = 22
striker_initial_pos = [300, 517]
player_score = 0
count = 0
friction_coefficient = 0.95
striker_speed = 10
aiming_angle = 0
speed_lower_limit = False
speed_higher_limit = True
final_text = ''

ball_list = [
    {
        "type": "queen",
        "color": "brown",
        "pos": [half_w, half_h],
        "vel": [0, 0],
        "radius": c_rad,
    },
    {
        "type": "black",
        "color": "grey",
        "pos": [half_w - c_rad, half_h - c_rad * math.sqrt(3)],
        "vel": [0, 0],
        "radius": c_rad,
    },
    {
        "type": "black",
        "color": "grey",
        "pos": [half_w + 2 * c_rad, half_h],
        "vel": [0, 0],
        "radius": c_rad,
    },
    {
        "type": "black",
        "color": "grey",
        "pos": [half_w - c_rad, half_h + c_rad * math.sqrt(3)],
        "vel": [0, 0],
        "radius": c_rad,
    },
    {
        "type": "black",
        "color": "grey",
        "pos": [half_w - 3 * c_rad, half_h - c_rad * math.sqrt(3)],
        "vel": [0, 0],
        "radius": c_rad,
    },
    {
        "type": "black",
        "color": "grey",
        "pos": [half_w, half_h - 2 * c_rad * math.sqrt(3)],
        "vel": [0, 0],
        "radius": c_rad,
    },
    {
        "type": "black",
        "color": "grey",
        "pos": [half_w + 3 * c_rad, half_h - c_rad * math.sqrt(3)],
        "vel": [0, 0],
        "radius": c_rad,
    },
    {
        "type": "black",
        "color": "grey",
        "pos": [half_w + 3 * c_rad, half_h + c_rad * math.sqrt(3)],
        "vel": [0, 0],
        "radius": c_rad,
    },
    {
        "type": "black",
        "color": "grey",
        "pos": [half_w, half_h + 2 * c_rad * math.sqrt(3)],
        "vel": [0, 0],
        "radius": c_rad,
    },
    {
        "type": "black",
        "color": "grey",
        "pos": [half_w - 3 * c_rad, half_h + c_rad * math.sqrt(3)],
        "vel": [0, 0],
        "radius": c_rad,
    },
    {
        "type": "white",
        "color": "white",
        "pos": [half_w - 2 * c_rad, half_h],
        "vel": [0, 0],
        "radius": c_rad,
    },
    {
        "type": "white",
        "color": "white",
        "pos": [half_w + c_rad, half_h - c_rad * math.sqrt(3)],
        "vel": [0, 0],
        "radius": c_rad,
    },
    {
        "type": "white",
        "color": "white",
        "pos": [half_w + c_rad, half_h + c_rad * math.sqrt(3)],
        "vel": [0, 0],
        "radius": c_rad,
    },
    {
        "type": "white",
        "color": "white",
        "pos": [half_w - 4 * c_rad, half_h],
        "vel": [0, 0],
        "radius": c_rad,
    },
    {
        "type": "white",
        "color": "white",
        "pos": [half_w - 2 * c_rad, half_h - 2 * c_rad * math.sqrt(3)],
        "vel": [0, 0],
        "radius": c_rad,
    },
    {
        "type": "white",
        "color": "white",
        "pos": [half_w + 2 * c_rad, half_h - 2 * c_rad * math.sqrt(3)],
        "vel": [0, 0],
        "radius": c_rad,
    },
    {
        "type": "white",
        "color": "white",
        "pos": [half_w + 4 * c_rad, half_h],
        "vel": [0, 0],
        "radius": c_rad,
    },
    {
        "type": "white",
        "color": "white",
        "pos": [half_w + 2 * c_rad, half_h + 2 * c_rad * math.sqrt(3)],
        "vel": [0, 0],
        "radius": c_rad,
    },
    {
        "type": "white",
        "color": "white",
        "pos": [half_w - 2 * c_rad, half_h + 2 * c_rad * math.sqrt(3)],
        "vel": [0, 0],
        "radius": c_rad,
    },
    {
        "type": "striker",
        "color": "yellow",
        "pos": striker_initial_pos,
        "vel": [0, 0],
        "radius": striker_radius,
    },
]

def check_wall_collision():
    global carrom_width, carrom_height
    for i in ball_list:
        if i["pos"][0] - i["radius"] < 30 or i["pos"][0] + i["radius"] > carrom_width:
            i["vel"][0] = -i["vel"][0]  # Reverse the x-velocity for a wall collision
        if i["pos"][1] - i["radius"] < 30 or i["pos"][1] + i["radius"] > carrom_height:
            i["vel"][1] = -i["vel"][1]  # Reverse the y-velocity for a wall collision

def check_collisions():
    # Iterate through all pairs of balls
    for i in range(len(ball_list)):
        for j in range(i + 1, len(ball_list)):
            ball1 = ball_list[i]
            ball2 = ball_list[j]

            # Calculate the distance between the centers of the two balls
            dx = ball2["pos"][0] - ball1["pos"][0]
            dy = ball2["pos"][1] - ball1["pos"][1]
            distance = math.sqrt(dx ** 2 + dy ** 2)

            # Check if there is a collision between the two balls
            if distance < ball1["radius"] + ball2["radius"]:
                # Handle the collision
                handle_ball_collision(ball1, ball2)

def handle_ball_collision(ball1, ball2):
    pos1 = ball1["pos"]
    pos2 = ball2["pos"]
    vel1 = ball1["vel"]
    vel2 = ball2["vel"]
    radius1 = ball1["radius"]
    radius2 = ball2["radius"]

    dx = pos2[0] - pos1[0]
    dy = pos2[1] - pos1[1]
    distance = math.sqrt(dx ** 2 + dy ** 2)

    if distance < radius1 + radius2:
        # Calculate the collision normal
        collision_normal = [dx / distance, dy / distance]

        # Calculate relative velocity
        relative_velocity = [vel2[0] - vel1[0], vel2[1] - vel1[1]]

        # Calculate relative speed in the direction of the collision normal
        speed_along_normal = relative_velocity[0] * collision_normal[0] + relative_velocity[1] * collision_normal[1]

        # If they are moving away from each other, do nothing
        if speed_along_normal > 0:
            return

        # Calculate impulse (change in velocity)
        impulse_magnitude = -2 * speed_along_normal / (1 / radius1 + 1 / radius2)

        # Apply impulse to both balls
        ball1["vel"][0] -= (impulse_magnitude / radius1) * collision_normal[0]
        ball1["vel"][1] -= (impulse_magnitude / radius1) * collision_normal[1]
        ball2["vel"][0] += (impulse_magnitude / radius2) * collision_normal[0]
        ball2["vel"][1] += (impulse_magnitude / radius2) * collision_normal[1]

        # Move the balls to prevent overlap
        overlap = (radius1 + radius2 - distance) / 2
        pos1[0] -= overlap * collision_normal[0]
        pos1[1] -= overlap * collision_normal[1]
        pos2[0] += overlap * collision_normal[0]
        pos2[1] += overlap * collision_normal[1]

def check_pocket_conditions(ball):
    global pocket_coordinates, pocket_radius, player_score, final_text
    for pocket_x, pocket_y in pocket_coordinates:
        distance_to_pocket = math.sqrt(
            (ball["pos"][0] - pocket_x) ** 2 + (ball["pos"][1] - pocket_y) ** 2
        )
        if distance_to_pocket <= pocket_radius:
            # Ball has gone into a pocket, remove it from the game
            player_score += 1
            ball_list.remove(ball)
            if ball['type'] == 'striker' :
                player_score -= 1
                ball['pos'] = [300,517]
                ball['vel'] = [0,0]
                ball_list.append(ball)
            if ball_list[0]['type'] == 'striker' :
                final_text = 'Congratulation!! You Won!!'

def mouse_drag(pos):
    global ball_list
    if count == 0 :
        if (ball_list[-1]['pos'][0]>=124 and ball_list[-1]['pos'][0]<=476):
            ball_list[-1]['pos'][0] = pos[0]
        if (ball_list[-1]['pos'][0]<124):
            ball_list[-1]['pos'][0] = 124
        elif (ball_list[-1]['pos'][0]>476) :
            ball_list[-1]['pos'][0] = 476

def keydown(key):
    global ball_list, count, aiming_angle, striker_speed
    if count == 0 :
        if key == simplegui.KEY_MAP['left']:
            aiming_angle -= 0.01  # Adjust the angle increment as needed
        elif key == simplegui.KEY_MAP['right']:
            aiming_angle += 0.01  # Adjust the angle increment as needed
        
        if speed_lower_limit :
            if key == simplegui.KEY_MAP['down']:
                striker_speed -= 10
        if speed_higher_limit :
            if key == simplegui.KEY_MAP['up']:
                striker_speed += 10
    else :
        if key == simplegui.KEY_MAP['space']:
            ball_list[-1]['pos'] = [300,517]
            ball_list[-1]['vel'] = [0,0]
            count = 0

def keyup(key):
    global aiming_angle, striker_speed, count
    if count == 0 :
        if (key == simplegui.KEY_MAP['f'] or key == simplegui.KEY_MAP['F'] ) :
            ball_list[-1]['vel'][1] = -striker_speed * math.cos(aiming_angle)
            ball_list[-1]['vel'][0] = striker_speed * math.sin(aiming_angle)
            count += 1

# Draw_handler
def draw_handler(canvas):
    global speed_lower_limit, striker_speed, speed_higher_limit
    canvas.draw_image(
        image, (half_w, half_h), (width, height), (half_w, half_h), (width, height)
    )
    canvas.draw_text("Score: " + str(player_score), (20, 20), 20, "white")
    for ball in ball_list:
        canvas.draw_circle(ball["pos"], ball["radius"], 1, 'black', ball["color"])
    canvas.draw_text(final_text,(30,150),50,'white')
    
    for ball in ball_list:
        ball["pos"][0] += ball["vel"][0]
        ball["pos"][1] += ball["vel"][1]
        ball["vel"][0] *= friction_coefficient
        ball["vel"][1] *= friction_coefficient

    if (striker_speed>10) :
        speed_lower_limit = True
    else :
        speed_lower_limit = False

    if (striker_speed<40) :
        speed_higher_limit = True
    else :
        speed_higher_limit = False

    line_length = 50  # Adjust the length of the aiming line
    endpoint_x = ball_list[-1]['pos'][0] + line_length * (2 * aiming_angle)  # Change the angle factor as needed
    endpoint_y = ball_list[-1]['pos'][1] - 50
    canvas.draw_line(ball_list[-1]['pos'],[endpoint_x,endpoint_y],2,'black')
    
    meter_height = 200
    speed_meter = striker_speed*5
    canvas.draw_polygon([(10, 100), (20, 100), (20, 100 + meter_height), (10, 100 + meter_height)], 2, 'Black', 'White')
    canvas.draw_polygon([(10, 100 + meter_height - speed_meter), (20, 100 + meter_height - speed_meter),(20, 100 + meter_height), (10, 100 + meter_height)], 2, 'Black', 'Red')

    check_wall_collision()

    check_collisions()

    for i in ball_list :
        check_pocket_conditions(i)

frame = simplegui.create_frame("Carrom Game", width, height)
image = simplegui._load_local_image("carrom.png")
frame.set_draw_handler(draw_handler)
frame.set_keydown_handler(keydown)
frame.set_keyup_handler(keyup)
frame.set_mousedrag_handler(mouse_drag)
frame.add_label('Welcome to Carrom Game!')
frame.add_label('')
frame.add_label('Press Up Arrow to increase striker speed and Press Down Arrow to decrease striker speed.')
frame.add_label('')
frame.add_label('Use Left and Right arrow keys to aim.')
frame.add_label('')
frame.add_label('Use the Mouse to drag and place the striker.')
frame.add_label('')
frame.add_label('Press f key to shoot the striker.')
frame.add_label('')
frame.add_label('After every move you need to press Space to reset the striker.')
frame.add_label('')
frame.add_label('Remove all the balls to win the game.')

frame.start()
```
